```python
#importação de bibliotecas
import numpy as np
import pandas as pd
import seaborn as sns
 
```

```python
df = pd.read_csv("Pokemon.csv")
df
```

```python
sns.histplot(df["Defense"], bins =40, kde = True).set(title = "Gráfico de defesa")
```

```python
sns.set_theme(style="ticks", palette="pastel")
```

```python
sns.histplot(df["Defense"], bins =40, kde = True).set(title = "Gráfico de defesa")
```

```python
sns.boxplot(x="Defense", y="Type 1", data=df)
sns.despine(offset=10, trim=True)
```

```python
sns.boxplot(x="Type 1", y="Defense", native_scale=True, data=df)
sns.despine(offset=10, trim=True)
```

```python
sns.boxplot(x="Type 1", y="Defense", linewidth=2.8, native_scale=True, data=df)
sns.despine(offset=10, trim=True)
```

```python
sns.boxplot(data=df, x="Type 1", y="Defense",  hue="Type 1", whis=(0, 100))
```

```python
sns.boxplot(data=df, y="Type 1", x="Defense",  hue="Type 1", whis=(0, 100))
sns.despine(offset=20, trim=True)
```

```python
sns.boxplot(data=df, y="Type 1", x="Attack",  hue="Type 1", whis=(0, 100))
sns.despine(offset=20, trim=True)
```

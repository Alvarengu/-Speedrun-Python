```python
#importação de bibliotecas
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt 
```

```python
df = pd.read_csv("Pokemon.csv")
df
```

```python
df.sort_values(by=("Attack"))
```

```python
df.sort_values(by=("Attack"), ascending=False).head(10)
```

```python
higher_attack = df.sort_values(by=("Attack"), ascending=False).head(10)
```

```python
plt.figure(figsize=(10, 6), dpi=120)
sns.scatterplot(data=higher_attack, x="Attack", y="Name", hue="Legendary", s=150)
```

```python
plt.figure(figsize=(10, 6), dpi=120)
sns.scatterplot(data=higher_attack, x="Attack", y="Name", hue="Type 1", s=150)
```

```python
df.sort_values(by=("Attack"), ascending=False).tail(10)
```

```python
low_attack = df.sort_values(by=("Attack"), ascending=False).tail(10)
sns.scatterplot(data=low_attack, x="Attack", y="Name", hue="Type 1", s=150)
```

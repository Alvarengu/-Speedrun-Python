```python
#importação de bibliotecas
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt 
```

```python
df = pd.read_csv("Pokemon.csv")
df
```

```python
sns.lineplot(data=df, x="Generation", y="Defense", color="red")
```

```python
sns.boxplot(data=df, x="Attack")
```

```python
sns.scatterplot(data=df, x="Attack", y="Type 1", hue="Generation")
```

```python
fire_df = df[df["Type 1"] == "Fire"]
```

```python
sns.scatterplot(data=fire_df, x="Attack", y="Defense", hue="Generation")
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=fire_df, x="Attack", y="Defense", hue="Generation")
```

```python
plt.figure(figsize=(10, 6))
poison_df = df[df["Type 1"] == "Poison"]
sns.scatterplot(data=poison_df, x="Attack", y="Defense", hue="Generation")
```

```python
plt.figure(figsize=(10, 6), dpi=120)
ice_df = df[df["Type 1"] == "Ice"]
sns.scatterplot(data=ice_df, x="Attack", y="Defense", hue="Generation", s=150)
```

```python
#importação de bibliotecas
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt 
```

```python
df = pd.read_csv("Pokemon.csv")
df
```

```python
plt.figure(figsize=(10, 6), dpi=120)
ice_df = df[df["Type 1"] == "Ice"]
sns.scatterplot(data=ice_df, x="Attack", y="Defense", hue="Generation", s=150)
```

```python
plt.figure(figsize=(10, 6), dpi=120)
sns.scatterplot(df, x="Generation", y="HP")
```

```python
plt.figure(figsize=(10, 6), dpi=120)
sns.scatterplot(df, x="Generation", y="HP", hue="Legendary")
```
